```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("F:\\Data Analysts\\Data Analaysts Elavate Lab inter\\train.csv")
df
```

```python
df.info()
df.describe()
```

```python
df.value_counts()
```

```python
sns.pairplot(df)
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
# Compute correlation matrix
corr = numeric_df.corr()
# Plot heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Histograms of Numeric Features
df.select_dtypes(include='number').hist(bins=15, figsize=(12,8), layout=(2,4))
plt.suptitle('Histograms of Numeric Features')
plt.show()
```

```python
import math
numeric_cols = df.select_dtypes(include='number').columns
num_plots = len(numeric_cols)
ncols = 2
nrows = math.ceil(num_plots / ncols)  # Calculate rows needed
plt.figure(figsize=(12, 4 * nrows))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(nrows, ncols, i)  # Use dynamic rows and cols here
    sns.boxplot(y=col, data=df)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```
